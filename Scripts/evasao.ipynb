```python
import pandas as pd
```

```python
def taxa_de_evasao(df_atual, df_anterior):
    df_at  = pd.read_csv(df_atual,    sep=';', encoding='latin1', low_memory=False)
    df_ant = pd.read_csv(df_anterior, sep=';', encoding='latin1', low_memory=False)

    def somacols(df, cols):
        """Soma colunas que existam no df, ignora as ausentes."""
        return sum(df[c].sum() for c in cols if c in df.columns)

    # Ingressantes
    i_at   = somacols(df_at,  ["QT_ING"])
    ip_at  = somacols(df_at,  ["QT_ING_PROUNII", "QT_ING_PROUNIP"])
    if_at  = somacols(df_at,  ["QT_ING_FIES"])          # só FIES federal

    i_ant  = somacols(df_ant, ["QT_ING"])
    ip_ant = somacols(df_ant, ["QT_ING_PROUNII", "QT_ING_PROUNIP"])
    if_ant = somacols(df_ant, ["QT_ING_FIES"])

    # Matriculados
    m_at   = somacols(df_at,  ["QT_MAT"])
    mp_at  = somacols(df_at,  ["QT_MAT_PROUNII", "QT_MAT_PROUNIP"])
    mf_at  = somacols(df_at,  ["QT_MAT_FIES"])

    m_ant  = somacols(df_ant, ["QT_MAT"])
    mp_ant = somacols(df_ant, ["QT_MAT_PROUNII", "QT_MAT_PROUNIP"])
    mf_ant = somacols(df_ant, ["QT_MAT_FIES"])

    # Concluintes
    c_at   = somacols(df_at,  ["QT_CONC"])
    cp_at  = somacols(df_at,  ["QT_CONC_PROUNII", "QT_CONC_PROUNIP"])
    cf_at  = somacols(df_at,  ["QT_CONC_FIES"])

    c_ant = somacols(df_ant, ["QT_CONC"])
    cp_ant = somacols(df_ant, ["QT_CONC_PROUNII", "QT_CONC_PROUNIP"])
    cf_ant = somacols(df_ant, ["QT_CONC_FIES"])

    # Taxas de permanência e evasão
    te = 1 - (m_at - i_at) / (m_ant - c_ant)
    te_p = 1 - (mp_at - ip_at) / (mp_ant - cp_ant)
    te_f = 1 - (mf_at - if_at) / (mf_ant - cf_ant)

    return float(te), float(te_p), float(te_f)
```

```python
try:
    for ano in range(2015, 2025):
        print(f"Taxas de evasão para {ano}: {taxa_de_evasao(f'microdados_censo_da_educacao_superior/MICRODADOS_CADASTRO_CURSOS_{ano}.csv', f'microdados_censo_da_educacao_superior/MICRODADOS_CADASTRO_CURSOS_{ano-1}.csv')}")
except FileNotFoundError as e:
    print(f"Arquivo não encontrado: {e.filename}")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```
