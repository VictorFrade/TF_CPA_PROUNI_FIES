```python
def taxa_de_evasao(df_atual, df_anterior):
    df_at = pd.read_csv(df_atual, sep=';', encoding='latin1', low_memory=False)
    df_ant = pd.read_csv(df_anterior, sep=';', encoding='latin1', low_memory=False)

    # Ingressantes
    i_at = df_at["QT_ING"].sum()
    ip_at = df_at["QT_ING_PROUNIP"].sum() + df_at["QT_ING_PROUNII"].sum()
    if_at = df_at["QT_ING_FIES"].sum() + df_at["QT_ING_RPFIES"].sum() + df_at["QT_ING_NRPFIES"].sum()

    i_ant = df_ant["QT_ING"].sum()
    ip_ant = df_ant["QT_ING_PROUNIP"].sum() + df_ant["QT_ING_PROUNII"].sum()
    if_ant = df_ant["QT_ING_FIES"].sum() + df_ant["QT_ING_RPFIES"].sum() + df_ant["QT_ING_NRPFIES"].sum()

    # Matriculados
    m_at = df_at["QT_MAT"].sum()
    mp_at = df_at["QT_MAT_PROUNIP"].sum() + df_at["QT_MAT_PROUNII"].sum()
    mf_at = df_at["QT_MAT_FIES"].sum() + df_at["QT_MAT_RPFIES"].sum() + df_at["QT_MAT_NRPFIES"].sum()

    m_ant = df_ant["QT_MAT"].sum()
    mp_ant = df_ant["QT_MAT_PROUNIP"].sum() + df_ant["QT_MAT_PROUNII"].sum()
    mf_ant = df_ant["QT_MAT_FIES"].sum() + df_ant["QT_MAT_RPFIES"].sum() + df_ant["QT_MAT_NRPFIES"].sum()

    # Concluintes
    c_at = df_at["QT_CONC"].sum()
    cp_at = df_at["QT_CONC_PROUNIP"].sum() + df_at["QT_CONC_PROUNII"].sum()
    cf_at = df_at["QT_CONC_FIES"].sum() + df_at["QT_CONC_RPFIES"].sum() + df_at["QT_CONC_NRPFIES"].sum()

    c_ant = df_ant["QT_CONC"].sum()
    cp_ant = df_ant["QT_CONC_PROUNIP"].sum() + df_ant["QT_CONC_PROUNII"].sum()
    cf_ant = df_ant["QT_CONC_FIES"].sum() + df_ant["QT_CONC_RPFIES"].sum() + df_ant["QT_CONC_NRPFIES"].sum()

    perm = (c_at + m_at) / (i_ant + m_ant)
    perm_p = (cp_at + mp_at) / (ip_ant + mp_ant)
    perm_f = (cf_at + mf_at) / (if_ant + mf_ant)

    te = 1 - perm
    te_p = 1 - perm_p
    te_f = 1 - perm_f

    return (te, te_p, te_f)        
```
