```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def pizza(dados, titulo):
    """Gera um gráfico de pizza simples para a série de dados fornecida."""
    plt.figure(figsize=(10, 6))
    dados.value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
    plt.title(titulo)
    plt.ylabel('')
    plt.show()

def histograma(dados, titulo, x, y):
    """Gera um histograma simples para a série de dados fornecida."""
    plt.figure(figsize=(10, 6))
    sns.histplot(dados, bins=20, kde=True)
    plt.title(titulo)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

```python
prouni_df_2020["DATA_NASCIMENTO"].str.split("/").str[2].astype(int).describe()
# Incoerência na data de nascimento máxima
```

```python
ano_nascimento_2020 = prouni_df_2020["DATA_NASCIMENTO"].str.split('/').str[2].astype(int)
print("Média:", np.mean(ano_nascimento_2020).round())
print("Mediana:", np.median(ano_nascimento_2020))
print("Moda:", ano_nascimento_2020.mode()[0])

q1 = ano_nascimento_2020.quantile(0.25)
q3 = ano_nascimento_2020.quantile(0.75)
iqr = q3 - q1
print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)

print("Idade mínima:", 2020 - ano_nascimento_2020.max())
print("Idade máxima:", 2020 - ano_nascimento_2020.min())
```

```python
prouni_20_limpo = prouni_df_2020.copy()
prouni_20_limpo["ANO_NASCIMENTO"] = prouni_20_limpo["DATA_NASCIMENTO"].str.split('/').str[2].astype(int)
prouni_20_limpo = prouni_20_limpo[(prouni_20_limpo["ANO_NASCIMENTO"] >= 1920) & (prouni_20_limpo["ANO_NASCIMENTO"] <= 2005)]
ano_nasc_2020_limpo = prouni_20_limpo["ANO_NASCIMENTO"]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=ano_nasc_2020_limpo)
plt.title('Boxplot dos Anos de Nascimento dos Beneficiários do Prouni em 2020')
plt.xlabel('Ano de Nascimento')
plt.show()
```
